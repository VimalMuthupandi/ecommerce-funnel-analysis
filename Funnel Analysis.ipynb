```python
# Import all the libraries we need
import pandas as pd               # For data manipulation (like Excel but in Python)
import matplotlib.pyplot as plt   # For creating charts and plots
import matplotlib.ticker as mticker  # For formatting axis numbers (e.g. 1000 → 1,000)
import seaborn as sns             # Built on matplotlib — makes charts look better
import warnings
warnings.filterwarnings('ignore') # Suppress unnecessary warning messages
import matplotlib.patches as mpatches


print("✅ Libraries imported successfully!")
```

```python
# Read each CSV file into a pandas DataFrame
orders      = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_orders_dataset.csv')
geoloc      = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_geolocation_dataset.csv')
customers   = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_customers_dataset.csv')
items       = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_order_items_dataset.csv')
payments    = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_order_payments_dataset.csv')
reviews     = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_order_reviews_dataset.csv')
products    = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_products_dataset.csv')
sellers     = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/olist_sellers_dataset.csv')
cat_trans   = pd.read_csv('/Users/vimalmuthupandi/Desktop/Brazillian E-Commerce Data/product_category_name_translation.csv')

print("✅ All datasets loaded!")

# Quick check — see shape (rows, columns) of each file
print(f"orders    : {orders.shape}")
print(f"geoloc    : {geoloc.shape}")
print(f"customers : {customers.shape}")
print(f"items     : {items.shape}")
print(f"payments  : {payments.shape}")
print(f"reviews   : {reviews.shape}")
print(f"products  : {products.shape}")
print(f"sellers   : {sellers.shape}")
print(f"cat_trans : {cat_trans.shape}")
```

```python
# These columns are loaded as plain text (strings) by default
date_cols = [
    'order_purchase_timestamp',       # When customer placed the order
    'order_approved_at',              # When payment was approved
    'order_delivered_carrier_date',   # When seller handed to courier
    'order_delivered_customer_date',  # When customer received it
    'order_estimated_delivery_date'   # Estimated delivery date shown to customer
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col]) # We convert them to datetime so we can do date math (e.g. subtract dates)

# Verify conversion worked
print("✅ Date columns converted!")
print(orders[date_cols].dtypes) # dtypes - Data Types
```

```python
# --- Products: fill missing category name with 'unknown' ---
# We don't drop rows because we'd lose product data
products['product_category_name'] = products['product_category_name'].fillna('unknown')

# --- Orders: missing delivery dates are NOT errors ---
# A "shipped" order won't have a delivery date yet — that's expected
# So we leave those nulls as-is

# --- Reviews: missing comment title/message is normal ---
# Most users only give a star rating without writing text — we keep as-is

print("✅ Missing values handled!")
print(f"Products null count after fix: {products['product_category_name'].isnull().sum()}")
```

```python
# The products table has category names in Portuguese
# We merge the translation table to get English names

# 'how=left' means: keep ALL rows from products
# even if there's no matching translation (those will get NaN)
products = products.merge(cat_trans, on='product_category_name', how='left')

# Fill any remaining untranslated categories
products['product_category_name_english'] = (
    products['product_category_name_english'].fillna('unknown')
)

print("✅ English category names added!")
display(products[['product_category_name', 'product_category_name_english']].head(10))
```

```python
# Extract Year-Month for time series (e.g. "2018-03")
# Useful for grouping orders by month later
orders['order_year_month'] = orders['order_purchase_timestamp'].dt.to_period('M')

# Calculate how many days each order took to be delivered
# datetime subtraction gives a timedelta, .dt.days converts it to an integer
orders['delivery_days'] = (
    orders['order_delivered_customer_date'] - orders['order_purchase_timestamp']
).dt.days

# Create a True/False flag: was the delivery late?
# Late = actual delivery date is AFTER the estimated delivery date
orders['is_late'] = (
    orders['order_delivered_customer_date'] > orders['order_estimated_delivery_date']
)

print("✅ New columns added!")
display(orders[['order_id', 'order_status', 'order_year_month',
                'delivery_days', 'is_late']].head())
```

```python
# --- Aggregate Order Items ---
# One order can have multiple items, so we group by order_id
# and calculate: total items count, total price, total freight
items_agg = items.groupby('order_id').agg(
    total_items   = ('order_item_id', 'count'),  # count of items in order
    total_price   = ('price', 'sum'),             # sum of all item prices
    total_freight = ('freight_value', 'sum')      # sum of all freight costs
).reset_index()  # converts the groupby index back to a normal column

# --- Aggregate Payments ---
# One order can have multiple payment rows (e.g. credit card + voucher)
payments_agg = payments.groupby('order_id').agg(
    total_payment = ('payment_value', 'sum'),
    payment_type  = ('payment_type', lambda x: x.mode().iloc[0] if not x.mode().empty else 'Unknown'
)  # most used payment method
).reset_index()

# --- Aggregate Reviews ---
# Some orders have multiple reviews — we take the average score
reviews_agg = reviews.groupby('order_id').agg(
    avg_review_score = ('review_score', 'mean')
).reset_index()

# --- Merge Everything into One Master Table ---
# We start from orders and LEFT JOIN each table on order_id or customer_id
# LEFT JOIN = keep all rows from the left table even if no match on the right
master = (
    orders
    .merge(customers,    on='customer_id', how='left')
    .merge(items_agg,    on='order_id',    how='left')
    .merge(payments_agg, on='order_id',    how='left')
    .merge(reviews_agg,  on='order_id',    how='left')
)

print("✅ Master table created!")
print(f"Shape: {master.shape}")
display(master.head())

# After groupby, order_id becomes an index.
# reset_index() converts it back into a normal column.
# Without this, merging becomes messy.
```

```python
# Filter only delivered orders for delivery-related stats
delivered = master[master['order_status'] == 'delivered']

print("📊 BUSINESS SUMMARY STATS")

print(f"Total Orders          : {len(master):,}")
print(f"Delivered Orders      : {len(delivered):,}  ({len(delivered)/len(master)*100:.1f}%)")
print(f"Cancelled Orders      : {(master['order_status']=='canceled').sum():,}")
print(f"Avg Delivery Days     : {delivered['delivery_days'].mean():.1f} days")
print(f"Late Deliveries       : {master['is_late'].sum():,}  ({master['is_late'].mean()*100:.1f}%)")
print(f"Avg Review Score      : {master['avg_review_score'].mean():.2f} / 5")
print(f"Avg Order Value       : R$ {master['total_payment'].mean():.2f}")
print(f"Total Revenue         : R$ {master['total_payment'].sum():,.2f}")
```

```python
# Each stage of the funnel is defined by a condition on the orders table
# We count how many orders REACHED each stage

funnel_stages = {
    'Order Placed'          : len(orders),
    'Payment Approved'      : orders['order_approved_at'].notna().sum(),
    'Picked up by Carrier'  : orders['order_delivered_carrier_date'].notna().sum(),
    'Delivered to Customer' : orders['order_delivered_customer_date'].notna().sum(),
    'Review Given'          : orders['order_id'].isin(reviews['order_id']).sum()
}

# Convert to a DataFrame for easy use
funnel_df = pd.DataFrame(
    list(funnel_stages.items()),
    columns=['stage', 'order_count']
)

# Drop-off = how many orders were lost going from previous stage to this one
funnel_df['drop_off']    = funnel_df['order_count'].shift(1) - funnel_df['order_count']

# Drop-off rate = drop_off as % of the PREVIOUS stage
funnel_df['drop_off_pct'] = (funnel_df['drop_off'] / funnel_df['order_count'].shift(1) * 100).round(2)

# Conversion rate = how many % of the FIRST stage (orders placed) reached this stage
funnel_df['conversion_pct'] = (funnel_df['order_count'] / funnel_df['order_count'].iloc[0] * 100).round(2)

print("✅ Funnel table ready!")
display(funnel_df)


#len(orders) → total rows = total orders placed
#.notna().sum() → count rows where the date column is not empty (meaning that event happened)
#.isin(reviews['order_id']).sum() → count orders that exist in the reviews table (meaning a review was submitted)
```

```python
# Understanding HOW LONG each step takes helps identify bottlenecks

orders['time_to_approval_hrs'] = (
    orders['order_approved_at'] - orders['order_purchase_timestamp']
).dt.total_seconds() / 3600   # convert seconds → hours

orders['time_approval_to_carrier_days'] = (
    orders['order_delivered_carrier_date'] - orders['order_approved_at']
).dt.days

orders['time_carrier_to_customer_days'] = (
    orders['order_delivered_customer_date'] - orders['order_delivered_carrier_date']
).dt.days

# Summary
print("=" * 45)
print("       ⏱️  TIME BETWEEN FUNNEL STAGES")
print("=" * 45)
print(f"Order → Approval       : {orders['time_to_approval_hrs'].mean():.1f} hrs avg")
print(f"Approval → Carrier     : {orders['time_approval_to_carrier_days'].mean():.1f} days avg")
print(f"Carrier → Customer     : {orders['time_carrier_to_customer_days'].mean():.1f} days avg")
print("=" * 45)
```

```python
fig, ax = plt.subplots(figsize=(13, 6))

colors = ['#1a73e8', '#34a853', '#fbbc04', '#ea4335', '#9c27b0']

bars = ax.bar(funnel_df['stage'], funnel_df['order_count'],
              color=colors, edgecolor='white', width=0.55)

ax.set_title('E-Commerce Purchase Funnel — Order Journey',
             fontsize=14, fontweight='bold')
ax.set_ylabel('Number of Orders')
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=20)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

# Label each bar with: order count + conversion % from start
for bar, row in zip(bars, funnel_df.itertuples()):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 400,
            f'{row.order_count:,}\n({row.conversion_pct:.1f}%)',
            ha='center', fontsize=9, fontweight='bold')

# Annotate drop-off arrows between bars
for i in range(1, len(funnel_df)):
    if pd.notna(funnel_df['drop_off'].iloc[i]):
        ax.annotate(
            f'▼ {funnel_df["drop_off"].iloc[i]:,.0f} lost\n({funnel_df["drop_off_pct"].iloc[i]:.1f}%)',
            xy=(i - 0.5, (funnel_df['order_count'].iloc[i] + funnel_df['order_count'].iloc[i-1]) / 2),
            fontsize=8, color='red', ha='center',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='#ffe0e0', edgecolor='red', alpha=0.8)
        )

plt.tight_layout()
plt.savefig('funnel_main.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Main funnel chart saved!")
```

```python
# This chart shows ONLY the drop-off % at each transition
# Easier to see WHERE the biggest leakage is

drop_df = funnel_df.dropna(subset=['drop_off_pct']).copy()

# Color bars: red if drop-off > 2%, orange if > 0.5%, green otherwise
bar_colors = [
    '#d9534f' if x > 2 else '#f0ad4e' if x > 0.5 else '#2e7d32'
    for x in drop_df['drop_off_pct']
]

fig, ax = plt.subplots(figsize=(11, 5))

bars = ax.bar(drop_df['stage'], drop_df['drop_off_pct'],
              color=bar_colors, edgecolor='white', width=0.5)

ax.set_title('Funnel Drop-off Rate at Each Stage (%)',
             fontsize=14, fontweight='bold')
ax.set_ylabel('Drop-off Rate (%)')
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=20)

for bar, val, lost in zip(bars, drop_df['drop_off_pct'], drop_df['drop_off']):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.05,
            f'{val:.2f}%\n({lost:,.0f} orders)',
            ha='center', fontsize=9, fontweight='bold')

legend_e = [
    mpatches.Patch(color='#d9534f', label='Drop-off > 2% → Critical ❌'),
    mpatches.Patch(color='#f0ad4e', label='Drop-off 0.5–2% → Monitor ⚠️'),
    mpatches.Patch(color='#2e7d32', label='Drop-off < 0.5% → Healthy ✅'),
]
ax.legend(handles=legend_e, fontsize=9, frameon=True)

plt.tight_layout()
plt.savefig('funnel_dropoff.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Drop-off chart saved!")
```

```python
# Do customers who pay with credit card convert better than boleto?
# This segments the funnel conversion rate by payment method

payment_funnel = (payments
    .merge(orders[['order_id', 'order_status',
                   'order_delivered_customer_date']], on='order_id', how='left'))

pay_summary = payment_funnel.groupby('payment_type').agg(
    total_orders    = ('order_id', 'count'),
    delivered_orders = ('order_delivered_customer_date',
                        lambda x: x.notna().sum())
).reset_index()

# Conversion rate per payment type = delivered / total
pay_summary['conversion_rate'] = (
    pay_summary['delivered_orders'] / pay_summary['total_orders'] * 100
).round(2)

# Remove 'not_defined' (only 3 rows — not meaningful)
pay_summary = pay_summary[pay_summary['payment_type'] != 'not_defined']
pay_summary = pay_summary.sort_values('conversion_rate', ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Funnel Analysis — Payment Type Breakdown',
             fontsize=13, fontweight='bold')

# Left: Total orders per payment type
axes[0].barh(pay_summary['payment_type'], pay_summary['total_orders'],
             color=sns.color_palette('muted', len(pay_summary)), edgecolor='white')
axes[0].set_title('Total Orders by Payment Type')
axes[0].set_xlabel('Total Orders')
for i, v in enumerate(pay_summary['total_orders']):
    axes[0].text(v + 100, i, f'{v:,}', va='center', fontsize=9)

# Right: Conversion rate per payment type
colors_pay = ['#2e7d32' if x >= 95 else '#f0ad4e' if x >= 90 else '#d9534f'
               for x in pay_summary['conversion_rate']]
axes[1].barh(pay_summary['payment_type'], pay_summary['conversion_rate'],
             color=colors_pay, edgecolor='white')
axes[1].set_title('Delivery Conversion Rate by Payment Type (%)')
axes[1].set_xlabel('Conversion Rate (%)')
axes[1].set_xlim(80, 100)
for i, v in enumerate(pay_summary['conversion_rate']):
    axes[1].text(v + 0.1, i, f'{v:.1f}%', va='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('funnel_payment.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Payment funnel chart saved!")
```

```python
# Where and why are orders getting cancelled?
# We look at cancellations by state and by month

canceled = orders[orders['order_status'] == 'canceled'].copy()
canceled = canceled.merge(customers[['customer_id', 'customer_state']],
                          on='customer_id', how='left')

canceled['order_year_month'] = (
    canceled['order_purchase_timestamp'].dt.to_period('M').astype(str)
)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Cancellation Analysis', fontsize=13, fontweight='bold')

# Left: Top 10 states by cancellations
cancel_state = canceled['customer_state'].value_counts().head(10)
axes[0].barh(cancel_state.index[::-1], cancel_state.values[::-1],
             color='#d9534f', edgecolor='white')
axes[0].set_title('Top 10 States by Cancellations')
axes[0].set_xlabel('Cancelled Orders')
for i, v in enumerate(cancel_state.values[::-1]):
    axes[0].text(v + 0.5, i, str(v), va='center', fontsize=9)

# Right: Monthly cancellation trend
cancel_monthly = canceled.groupby('order_year_month').size().reset_index(name='cancellations')
axes[1].plot(cancel_monthly['order_year_month'],
             cancel_monthly['cancellations'],
             marker='o', color='#d9534f', linewidth=2)
axes[1].set_title('Monthly Cancellation Trend')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Cancellations')
axes[1].tick_params(axis='x', rotation=45)
axes[1].xaxis.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('funnel_cancellations.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Cancellation chart saved!")
```

```python
# This shows HOW LONG each step takes on average
# The longest step = the biggest bottleneck in the funnel

stage_times = {
    'Order →\nApproval (hrs)'      : orders['time_to_approval_hrs'].dropna(),
    'Approval →\nCarrier (days)'    : orders['time_approval_to_carrier_days'].dropna(),
    'Carrier →\nCustomer (days)'    : orders['time_carrier_to_customer_days'].dropna(),
}

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Funnel Bottleneck Analysis — Time Between Each Stage',
             fontsize=13, fontweight='bold')

colors_t = ['#1a73e8', '#fbbc04', '#ea4335']

for ax, (label, data), color in zip(axes, stage_times.items(), colors_t):
    # Cap extreme outliers for cleaner visuals
    data_capped = data[data <= data.quantile(0.98)]
    ax.hist(data_capped, bins=35, color=color, edgecolor='white')
    ax.axvline(data.mean(), color='black', linestyle='--',
               label=f'Mean: {data.mean():.1f}')
    ax.axvline(data.median(), color='gray', linestyle=':',
               label=f'Median: {data.median():.1f}')
    ax.set_title(label, fontsize=11, fontweight='bold')
    ax.set_ylabel('Number of Orders')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('funnel_bottleneck.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Bottleneck chart saved!")
```
